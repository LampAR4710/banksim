```python
import random


# =========================
# Q1. Character í´ë˜ìŠ¤
# =========================
class Character:
    def __init__(self, name, level, hp, attack, defense):
        self.name = name
        self.level = level
        self.hp = hp
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.hp > 0

    def take_damage(self, damage):
        real_damage = damage - self.defense
        if real_damage > 0:
            self.hp -= real_damage
        if self.hp < 0:
            self.hp = 0

    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        target.take_damage(damage)
        return damage


# =========================
# Q2. Player í´ë˜ìŠ¤
# =========================
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, attack=25, defense=5)
        self.experience = 0

    def gain_experience(self, exp):
        self.experience += exp

    def level_up(self):
        while self.experience >= 50:
            self.experience -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5
            print(f"â­ ë ˆë²¨ ì—…! í˜„ì¬ ë ˆë²¨: {self.level}")


# =========================
# Q2. Monster í´ë˜ìŠ¤
# =========================
class Monster(Character):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, hp, attack, defense)


# =========================
# Q3. battle í•¨ìˆ˜
# =========================
def battle(player, monster):
    print(f"\nâš”ï¸ {player.name} VS {monster.name} (Lv.{monster.level})")

    while player.is_alive() and monster.is_alive():
        dmg = player.attack_target(monster)
        print(f"{player.name}ì´(ê°€) {monster.name}ì—ê²Œ {dmg} ë°ë¯¸ì§€!")
        print(f"{monster.name} HP: {monster.hp}")

        if not monster.is_alive():
            break

        dmg = monster.attack_target(player)
        print(f"{monster.name}ì´(ê°€) {player.name}ì—ê²Œ {dmg} ë°ë¯¸ì§€!")
        print(f"{player.name} HP: {player.hp}")

    if player.is_alive():
        exp = monster.level * 20
        player.gain_experience(exp)
        player.level_up()
        print("ğŸ‰ ì „íˆ¬ ìŠ¹ë¦¬!")
        return True
    else:
        print("ğŸ’€ ì „íˆ¬ íŒ¨ë°°..")
        return False


# =========================
# Q4. main í•¨ìˆ˜
# =========================
def main():
    monster_dict = {
        "ìŠ¬ë¼ì„": 1,
        "ê³ ë¸”ë¦°": 2,
        "ì˜¤í¬": 3
    }

    name = input("í”Œë ˆì´ì–´ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
    player = Player(name)

    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)
        win = battle(player, monster)

        if not win:
            print("â˜ ï¸ ê²Œì„ì˜¤ë²„")
            return

    print("\nğŸ† ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ì²˜ì¹˜í–ˆìŠµë‹ˆë‹¤! ê²Œì„ í´ë¦¬ì–´!")


# =========================
# í”„ë¡œê·¸ë¨ ì‹¤í–‰
# =========================
main()
```
